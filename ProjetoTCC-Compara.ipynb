### 1. Definição do problema <br>
#### 1.1 Contextualização <br>
<p> O ano de 2020 marcou o início da pandemia de COVID-19, causada pelo coronavírus (SARS-CoV-2) 
<br><br> É patente que a pandemia produz repercussões de ordem biomédica e epidemiológica em escala global. No entanto, também há fortes repercussões e impactos nas áreas social, econômica, política, ambiental e outras, sem precedentes na história recente das epidemias.
<br><br> Segundo a Organização para a Cooperação e Desenvolvimento Econômico (OCDE), em razão do surto de coronavírus, a economia global pode crescer na taxa mais baixa desde 2009.
<br><br> Uma das áreas da economia que sofreu forte impacto é o comércio exterior, em função da redução da atividade econômica em escala global.
<br><br> O Brasil, como todo país inserido numa economia globalizada, também foi atingido pelos efeitos da economia, principalmente por ter a China como principal parceiro econômico.    
</p>

#### 1.2 O problema proposto <br>
 O objetivo deste trabalho é apresentar uma análise sobre as importações nos anos de 2019 (antes da pandemia) e 2020 (durante a pandemia) e projetar o resultado dessa atividade para 2021. <br> 
> Why? <br>
>>  Uma das áreas da economia que sofreu forte impacto da pandemia de COVID-19 foi o comércio exterior, em função da redução da atividade econômica em escala global. A importação de mercadorias é importante indicador do estado da atividade econômica de um país e compreender seu comportamento pode ajudar na definição de políticas que favoreçam a retomada do crescimento econômico nacional.<br>

> Who? <br>
>> Os dados foram disponibilizados pelo Ministério da Indústria, Comércio Exterior e Serviços do Brasil. <br> 

> What? <br>
>> O objetivo dessa análise é tão somente conhecer como foi a repercussão imediata da pandemia nas importações realizadas pelo Brasil e projetar o comportamento esperado para o ano de 2021. <br>

> Where? <br>
>> A análise envolverá os países, os valores de importação e as mercadorias comercializadas. <br>

> When? <br>
>> Por se tratar do impacto imediato, analisaremos o período de janeiro/2019 a junho/2021.<br>

    
<br><br> Para isso, primeiramente, serão recuperados dados de importação de mercadorias referentes aos períodos em questão.
<br><br> Por fim, far-se-á a comparação entre os valores mês a mês para verificar como se deu o impacto da pandemia.

```python
#************************CÉLULA DE COLETA DE DADOS *****************************
import pandas as pd
import ssl #para desabilitar a verificação de segurança SSL. 
import os
from pathlib import Path

#Função para aquisição dos datasets
def get_datasets(urls_Datasets):
    
    #Desabilita verificação SSL
    ssl._create_default_https_context = ssl._create_unverified_context

    for url, destino in urls_Datasets.items(): #range(0,len(lista_datasets)):
        
        if not Path.exists(Path(destino)):
            try:
                df_temp = pd.read_csv(url, sep=';', encoding = 'cp1250')
                df_temp.to_csv(destino, sep=';',encoding = 'cp1250')
            except IOError:
                print("Um problema ocorreu processando %s ", destino)
                raise
#Fim da função

#Dicionário contendo as urls a serem acessadas e os arquivos de destino dos dados coletados
urls_Datasets = {
                 'https://balanca.economia.gov.br/balanca/bd/tabelas/PAIS.csv': 'Datasets/pais.csv',
                 'https://balanca.economia.gov.br/balanca/bd/tabelas/UF.csv'  : 'Datasets/uf.csv',
                 'https://balanca.economia.gov.br/balanca/bd/tabelas/NCM.csv' : 'Datasets/ncm.csv',
                 'https://balanca.economia.gov.br/balanca/bd/comexstat-bd/ncmv2/IMP_2019_V2.csv':'Datasets/importacoes2019.csv',
                 'https://balanca.economia.gov.br/balanca/bd/comexstat-bd/ncmv2/IMP_2020_V2.csv':'Datasets/importacoes2020.csv',
                 'https://balanca.economia.gov.br/balanca/bd/comexstat-bd/ncmv2/IMP_2021_V2.csv':'Datasets/importacoes2021.csv'

                }

get_datasets(urls_Datasets)    
```

```python
#***********Lendo as informações do dataset de importações dos anos 2019 a jun/2021***********

df_imp2019 = pd.read_csv('Datasets/importacoes2019.csv', sep=';', encoding='cp1250')
df_imp2019.info()

df_imp2020 = pd.read_csv('Datasets/importacoes2020.csv', sep=';', encoding='cp1250')
df_imp2020.info()

df_imp2021 = pd.read_csv('Datasets/importacoes2021.csv', sep=';', encoding='cp1250')
df_imp2021.info()
```

```python
#Concatenando os datasets
df_importacoes = pd.concat([df_imp2019, df_imp2020,df_imp2021])

df_importacoes.reset_index(drop=True, inplace=True) #Para reorganização dos índices
print(df_importacoes.shape)
df_importacoes.info()
```

```python
#Excluindo colunas de dataframe de importações
if set(['CO_UNID', 'CO_URF','CO_VIA']).issubset(df_importacoes.columns):
    df_importacoes.drop(['CO_UNID', 'CO_URF','CO_VIA'], axis=1, inplace=True)

#Convertando valores KG_LIQUIDO, VL_FOB, VL_FRETE e VL_SEGURO para float
df_importacoes['VL_FOB'] = df_importacoes['VL_FOB'].astype(float)
df_importacoes['VL_FRETE'] = df_importacoes['VL_FRETE'].astype(float)
df_importacoes['VL_SEGURO'] = df_importacoes['VL_SEGURO'].astype(float)
df_importacoes['KG_LIQUIDO'] = df_importacoes['KG_LIQUIDO'].astype(float)
df_importacoes.dtypes  #verificando se os tipos foram alterados
```

```python
#Deletando valores zerados para QT_ESTAT e VL_FOB 
#Verificando quantidade de valores zerados 
num_zeros = df_importacoes[(df_importacoes['QT_ESTAT'] == 0) &
              (df_importacoes['VL_FOB'] == 0)].count()

#Deletando se houver zero nos campos
if num_zeros.sum() > 0:
    df_importacoes.drop(df_importacoes.index[df_importacoes['QT_ESTAT'] == 0], inplace=True)
    df_importacoes.drop(df_importacoes.index[df_importacoes['VL_FOB'] == 0], inplace=True)
```

```python
#Criando a coluna VL_UNIT, obtida pela fórmula
# (VL_FOB+VL_FRETE+VL_SEGURO)/QT_ESTAT
df_importacoes.eval("VAL_UNIT = (VL_FOB+VL_FRETE+VL_SEGURO)/QT_ESTAT", target = df_importacoes, inplace = True)

df_importacoes.sample(n=10)
```

```python
#Converter preços mínimos menores que 1 em preço máximo
df_precos.loc[(df_precos['preco_minimo'] < 1 ) ,'preco_minimo'] = df_precos.preco_maximo
df_precos.head()
```

```python
#df_precos.info()
df_precos.eval("preco_medio = (preco_minimo+preco_maximo)/2", target = df_precos, inplace = True)
df_precos.describe() # estatísticas do dataframe


df_precos = df_precos.sort_values(by='preco_medio').reset_index(drop=True)
p25 = df_precos["preco_medio"].quantile(0.25)
p75 = df_precos['preco_medio'].quantile(0.75)
iqr = p75 - p25
print("p25 = %s, p75 = %s, iqr = %s ", p25, p75, iqr)
```

```python
f = pd.melt(df_precos, value_vars=['preco_minimo','preco_maximo', 'preco_medio'])
g = sns.FacetGrid(f, col="variable",  col_wrap=3, sharex=False, sharey=False, height = 5)
g = g.map(sns.histplot, "value")
```
